# ADDING FIPS CODE FOR BETTER VISULIZATION

```python
usa_fips_code = pd.read_csv(input_path+"USA_FIPS_Code.csv")
usa_fips_code.head()
```

```python
def create_county_state(countyName,state):
    return countyName + " County, "+state
```

```python
road_accident_data.shape
```

```python
road_accident_data['countyState'] = create_county_state(road_accident_data.County,road_accident_data.State)
```

```python
usa_fips_code = usa_fips_code.rename(columns={"County Name/State Abbreviation": "countyState"})
usa_fips_code.head()
```

```python
usa_fips_code = usa_fips_code.drop(["Year","Labor Force","Employed","Unemployed","Unemployment Rate (%)"], axis=1)
usa_fips_code.head()
```

```python
# usa_fips_code.drop(['LAUS Code'],axis=1)
usa_fips_code.head()
```

```python
merged_df = pd.merge(road_accident_data, usa_fips_code, how="left",on="countyState")
```
